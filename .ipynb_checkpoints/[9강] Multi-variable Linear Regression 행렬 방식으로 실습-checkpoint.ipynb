# [9강] Multi-variable Linear Regression 행렬 방식으로 실습

```python
import tensorflow as tf
import matplotlib as mlt
```

```python
x_data = [[73., 80., 75.], [93., 88., 93.], [89., 91., 90.], [96., 98., 100.], [73., 66., 70.]] #x 데이터를 인스턴스끼리 배열로 묶는다!
y_data = [[152.], [185.], [180.], [196.], [142.]] #x데이터의 각 배열로 묶인 인스턴스들과 하나씩 대응된다.
```

```python
X = tf.placeholder(tf.float32, shape=[None, 3]) #한 인스턴스에 3개의 요인이 들어있으므로 shape은 3이다.
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
W = tf.Variable(tf.random_normal([3, 1]), name = 'weight') #한 인스턴스에 3개의 요인이 들어있으므로 W도 요인에 대응되게 3개씩 되게 함.
b = tf.Variable(tf.random_normal([1]), name = 'bias')
```

```python
hypothesis = tf.matmul(X, W) + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cost_val, hy_val, _ = sess.run(
    [cost, hypothesis, train], feed_dict = {X: x_data, Y: y_data})
    if step%10 == 0:
        print(step, "Cost: ", cost_val, "\nPrediction:\n", hy_val)
```
