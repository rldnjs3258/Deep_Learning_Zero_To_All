# [10강] TensorFlow로 파일에서 데이터 읽어오기

#### 파일에서 데이터 읽어오는 방법!

```python
import tensorflow as tf
import numpy as np #numpy 라이브러리를 이용
tf.set_random_seed(777)
```

```python
xy = np.loadtxt('C:/Users/JIHYE/Desktop/data-01-test-score.csv', delimiter=',', dtype=np.float32) #loadtxt로 파일을 불러온다. ','로 구분하며 데이터 타입도 알려 준다.
#C:/Users/JIHYE/Desktop/ 는 바탕화면에 있는 파일을 불러오는 경로이다.
#파일의 텍스트를 x, y의 2차원 배열로 읽어 온다.
x_data = xy[:,0:-1] #2차원 배열 속의 각 배열들 전체의 인덱스들 중 처음부터 마지막 인덱스 전까지 x_data에 넣는다.
y_data = xy[:, [-1]] #2차원 배열 속의 각 배열들 전체의 인덱스들 중 마지막 인덱스만 y_data에 넣는다.
```

```python
print(x_data.shape, x_data, len(x_data)) #x_data의 모양, x_data, 길이를 프린트한다.
print(y_data.shape, y_data) #y_data의 모양, y_data를 프린트한다.
```

#### 아래는 원래와 동일하다!

```python
X= tf.placeholder(tf.float32, shape=[None, 3]) #요인이 3개 이므로
Y= tf.placeholder(tf.float32, shape=[None, 1]) #결과가 1개 이므로
```

```python
W = tf.Variable(tf.random_normal([3,1]), name='weight') #요인이 3개 이므로
b = tf.Variable(tf.random_normal([1]), name='bias') #결과가 1개 이므로
```

```python
hypothesis = tf.matmul(X, W) + b #Hypothesis를 매트릭스를 이용해 만듦
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y)) #cost
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5) #GradientDescent로 cost를 깎아 내려감
train = optimizer.minimize(cost) #cost를 optimizer 시킴
```

```python
sess = tf.Session() #세션
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001): #학습
    cost_val, hy_val, _= sess.run(
    [cost, hypothesis, train],
    feed_dict = {X:x_data, Y:y_data})
    if step%10 == 0:
        print(step, "Cost: ", cost_val,
             "\nPrediction:\n", hy_val)
```

```python
#학습한 값을 바탕으로 다른 값들을 예상할 수 있다.
print("Your score will be ", sess.run(hypothesis, feed_dict = {X: [[100, 70, 101]]}))
print("Other scores will be ", sess.run(hypothesis, feed_dict = {X: [[60, 70, 110], [90, 100, 80]]}))
```
