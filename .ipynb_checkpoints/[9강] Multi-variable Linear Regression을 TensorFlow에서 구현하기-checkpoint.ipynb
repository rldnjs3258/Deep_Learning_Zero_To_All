# [9강] Multi-variable Linear Regression을 TensorFlow에서 구현하기

```python
import tensorflow as tf
import matplotlib as mlt
```

```python
x1_data = [73., 93., 89., 96., 73.] #데이터 셋
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]
y_data = [152., 185., 180., 196., 142]
```

```python
x1 = tf.placeholder(tf.float32) #x1, x2, x3에 feed_dict 값을 주기 위해 placeholder로 만듬
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)
```

```python
Y = tf.placeholder(tf.float32)
```

```python
w1 = tf.Variable(tf.random_normal([1]), name='weight1') #W 값을 머신이 자동으로 설정하게 Variable로 설정
w2 = tf.Variable(tf.random_normal([1]), name='weight2')
w3 = tf.Variable(tf.random_normal([1]), name='weight3')
b = tf.Variable(tf.random_normal([1]), name = 'bias') #B
```

### Multi-variable Linear Regression의 hypothesis!

```python
hypothesis = x1 * w1 + x2 * w2 + x3 * w3 + b #hypothesis로 만들기
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y)) #cost
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5) #GradientDescent로 cost 작게 깎아 내려가기
train = optimizer.minimize(cost) #train
```

```python
sess = tf.Session() #Session
sess.run(tf.global_variables_initializer()) #Variable을 이용했으므로 설정
```

```python
for step in range(2000): #학습
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train], #train 돌려서 학습시키기
                                  feed_dict = {x1:x1_data, x2:x2_data, x3:x3_data, Y:y_data})
    if step % 10 == 0:
        print(step, "Cost: ", cost_val, "\nPrediction:\n", hy_val) #cost와 hypothesis를 기록한 후 출력
```
